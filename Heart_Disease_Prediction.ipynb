```python

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Dense, Dropout
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import accuracy_score, classification_report
X = result[selected]
y = a['Heart Disease']

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Standardize features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)


X_train = np.expand_dims(X_train, axis=1)
X_test = np.expand_dims(X_test, axis=1)

# Build multiple GRU models for voting
num_models = 5
models = []

for _ in range(num_models):
    model = Sequential()
    model.add(GRU(64, input_shape=(X_train.shape[1], X_train.shape[2])))
    model.add(Dropout(0.2))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    models.append(model)

# Train each model
for model in models:
    model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test),verbose=0)

# Voting-based prediction
predictions = []
for model in models:
    predictions.append(model.predict(X_test))

# Ensemble predictions (simple majority voting)
ensemble_predictions = np.round(np.mean(predictions, axis=0))

# Evaluate performance
accuracy = accuracy_score(y_test, ensemble_predictions)
print("Ensemble Accuracy:", accuracy)

print(classification_report(y_test, ensemble_predictions))
```

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, Flatten, Dense, LSTM, GRU
from tensorflow.keras.utils import to_categorical
import pandas as pd
from sklearn.metrics import matthews_corrcoef, confusion_matrix, f1_score
import numpy as np

# Assuming `result` and `a` are predefined DataFrames
X = result[selected]
y = a['Heart Disease']

def create_cnn(input_shape):
    model = Sequential()
    model.add(Conv1D(64, 1, activation='relu', input_shape=input_shape))
    model.add(Flatten())
    model.add(Dense(2, activation='softmax'))
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

def create_lstm(input_shape):
    model = Sequential()
    model.add(LSTM(64, input_shape=input_shape))
    model.add(Dense(2, activation='softmax'))
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

def create_gru(input_shape):
    model = Sequential()
    model.add(GRU(64, input_shape=input_shape))
    model.add(Dense(2, activation='softmax'))
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Models to evaluate
models = {
    'CNN': create_cnn,
    'LSTM': create_lstm,
    'GRU': create_gru
}

# Test different split ratios
test_sizes = [0.2, 0.3]
metrics_results = {
    'MCC': {},
    'Specificity': {},
    'F1': {},
    'NPV': {},
    'FPR': {},
    'FNR': {}
}

for metric in metrics_results:
    for model_name in models:
        metrics_results[metric][model_name] = []

for model_name, create_model_func in models.items():
    for test_size in test_sizes:
        # Split the data
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=0)

        # One-hot encode the target
        y_train_encoded = to_categorical(y_train, num_classes=2)
        y_test_encoded = to_categorical(y_test, num_classes=2)

        # Preprocess data
        X_train_encoded = pd.get_dummies(X_train)
        X_test_encoded = pd.get_dummies(X_test)

        # Align columns after encoding
        X_train_encoded, X_test_encoded = X_train_encoded.align(X_test_encoded, join='left', axis=1, fill_value=0)

        if model_name == 'CNN':
            # Reshape for CNN: (batch_size, steps, features)
            X_train_reshaped = X_train_encoded.values.reshape(X_train_encoded.shape[0], X_train_encoded.shape[1], 1).astype('float32')
            X_test_reshaped = X_test_encoded.values.reshape(X_test_encoded.shape[0], X_test_encoded.shape[1], 1).astype('float32')
        else:  # LSTM or GRU
            # Reshape for LSTM/GRU: (batch_size, timesteps, features)
            X_train_reshaped = X_train_encoded.values.reshape(X_train_encoded.shape[0], 1, X_train_encoded.shape[1]).astype('float32')
            X_test_reshaped = X_test_encoded.values.reshape(X_test_encoded.shape[0], 1, X_test_encoded.shape[1]).astype('float32')

        model = create_model_func(input_shape=X_train_reshaped.shape[1:])
        model.fit(X_train_reshaped, y_train_encoded, epochs=10, verbose=0)

        # Predict on the test set
        y_pred_probs = model.predict(X_test_reshaped)
        y_pred = y_pred_probs.argmax(axis=1)

        # Calculate confusion matrix elements
        tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()

        # Calculate MCC
        mcc = matthews_corrcoef(y_test, y_pred)
        metrics_results['MCC'][model_name].append(mcc)

        # Calculate Specificity
        specificity = tn / (tn + fp)
        metrics_results['Specificity'][model_name].append(specificity)

        # Calculate F1-score
        f1 = f1_score(y_test, y_pred, average='binary')
        metrics_results['F1'][model_name].append(f1)

        # Calculate NPV (Negative Predictive Value)
        npv = tn / (tn + fn)
        metrics_results['NPV'][model_name].append(npv)

        # Calculate FPR (False Positive Rate)
        fpr = fp / (fp + tn)
        metrics_results['FPR'][model_name].append(fpr)

        # Calculate FNR (False Negative Rate)
        fnr = fn / (fn + tp)
        metrics_results['FNR'][model_name].append(fnr)

# Plot all metrics as bar graphs
fig, axes = plt.subplots(3, 2, figsize=(18, 18))
fig.suptitle("Model Performance Metrics")

bar_width = 0.2
index = np.arange(len(test_sizes))

metric_names = ['MCC', 'Specificity', 'F1', 'NPV', 'FPR', 'FNR']
metric_titles = ["Matthew's Correlation Coefficient (MCC)", "Specificity", "F1-score", "Negative Predictive Value (NPV)", "False Positive Rate (FPR)", "False Negative Rate (FNR)"]

for ax, metric, title in zip(axes.flatten(), metric_names, metric_titles):
    for i, model_name in enumerate(models.keys()):
        ax.bar(index + i * bar_width, metrics_results[metric][model_name], bar_width, label=model_name)
    ax.set_title(title)
    ax.set_xlabel('Test Size')
    ax.set_ylabel(metric)
    ax.set_xticks(index + bar_width / 2)
    ax.set_xticklabels(test_sizes)
    ax.legend()

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
